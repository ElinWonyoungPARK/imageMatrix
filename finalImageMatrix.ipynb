```python
import numpy as np
import matplotlib.pyplot as plt

vertical = 1080
horizontal = 1920
my_dpi = 100

intervals = [19,10,7,29]
colors = [
    [0.2, 0.9, 0.2],
    [0.3, 0.3, 0.3],
    [0.8, 0.4, 0.4],
    [0.5, 0.5, 0.2],
]   
baseImg1 = np.full((vertical,horizontal,3),(255,255,255))
```

```python
def createBaseImg(vertical, horizontal):
    imgMatrix = np.full((vertical,horizontal,3),(255,255,255))
    return imgMatrix

def cutImg(imgMatrix,n,interval):
    subImgMatrix = imgMatrix[:,n:n+interval]
    return subImgMatrix

def coloring(subImgMatrix,color):
     subImgMatrix = subImgMatrix * color
     return subImgMatrix

def drawPattern(imgMatrix,intervals,colors):
    n = 0

    _,horizontal,_ = imgMatrix.shape
    while (n<horizontal) :
        for interval, color in zip(intervals, colors):
            imgMatrix[:, n: n+interval] = coloring(cutImg(imgMatrix, n, interval),color)
        
            n = n + interval 
    return imgMatrix
```

```python
finalImg = drawPattern(baseImg1,intervals,colors)
plt.figure(figsize=(horizontal/my_dpi, vertical/my_dpi), dpi=my_dpi)
plt.imshow(finalImg)
plt.axis('off')
plt.savefig('C:/Users/User/OpenCv/image9.tiff',dpi=my_dpi)
```
